## Loops concepts using pattern (Traingle, Diamond, Butterfly):




```python
size=int(input("enter number"))
for i in range(1,(size+1)):
    for j in range(0,i):
        print("* ",end="")
    for k in range(size,i,-1):
        print("  ",end="")
    for l in range(size,i,-1):
        print("  ",end="")
    for m in range(0,i):
        print("* ",end="")
    print("")
for i in range(1,(size+1)):
    for j in range((size+1),i,-1):
        print("* ",end="")
    for k in range(1,i):
        print(" ", end="")
        if i>1:
            for l in range(i+3,i,-1):
                print(" ", end="")
    for j in range((size+1),i,-1):
        print("* ",end="")
    print("")
    
   
    
```

```python
size=int(input("Enter the Size of Diamond shape:"))
for i in range(1,(size+1)):
    for j in range((size+1),i,-1):
        print(" ", end="")
    for k in range(1, (i+1)):
        print("* ", end="")
    print("")
for i in range(1,(size+1)):
    for k in range(1, (i+1)):
        print(" ", end="")
    for j in range((size+1),i,-1):
        print("* ", end="")
    
    print("")
```
